```python
#A1
import librosa
import librosa.display
import matplotlib.pyplot as plt
filename = '/content/lab1_recording.wav'
y, sr = librosa.load(filename)
librosa.display.waveshow(y, sr=sr)
plt.title('Speech Signal')
plt.show()
```

```python
#A2
length_seconds = len(y) / sr
print(f"Length of the signal: {length_seconds:.2f} seconds")


magnitude_range = max(abs(y)) - min(abs(y))
print(f"Magnitude range of the signal: {magnitude_range:.2f}")


spoken_words_threshold = 0.5
spoken_words_indices = abs(y) > spoken_words_threshold
spoken_words_duration = sum(spoken_words_indices) / sr
print(f"Duration of spoken words: {spoken_words_duration:.2f} seconds")


silence_threshold = 0.1
silence_indices = abs(y) < silence_threshold
silence_duration = sum(silence_indices) / sr
print(f"Duration of silence: {silence_duration:.2f} seconds")
```

```python
#A3
import IPython.display as ipd

start_time = 1.25
end_time = 4

segment = y[int(start_time * sr):int(end_time * sr)]

ipd.Audio(segment, rate=sr)
```

```python
#A4
def play_segment(start_time, end_time):
    segment = y[int(start_time * sr):int(end_time * sr)]
    ipd.display(ipd.Audio(segment, rate=sr))

while True:
    try:
        start_time = float(input("Enter start time (in seconds): "))
        end_time = float(input("Enter end time (in seconds): "))
        play_segment(start_time, end_time)
    except ValueError:
        print("Invalid input. Please enter numeric values.")

    continue_playing = input("Do you want to play another segment? (yes/no): ").lower()
    if continue_playing != 'yes':
        break
```
